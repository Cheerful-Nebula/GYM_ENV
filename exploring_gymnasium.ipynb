```python
from stable_baselines3 import PPO  # or DQN, A2C, etc.
from my_helper_funcs import evaluate_model, plot_position_curves_with_note

import os

# Set up experiment settings
env_id = "LunarLander-v3"
model_checkpoints = [10000, 20000, 30000, 40000]  # Timesteps when model was saved
base_model_dir = os.path.join(env_id, "saved_models")
reward_shaper = None  # Add your reward shaper instance if applicable
position_indices = [0,1]  # Track only x-position (index 0 in MountainCar)

# Collect position stat curves
position_curves = []
reward_curves = []
model = PPO("MlpPolicy", env_id, verbose=0)

for steps in model_checkpoints:

    model.learn(total_timesteps=steps, reset_num_timesteps=False)

    avg_reward, avg_position_stats = evaluate_model(
        model,
        env_id=env_id,
        num_episodes=5,
        reward_shaper=reward_shaper,
        render=False,
        record_video=False,
        experiment_num=1,
        position_indices=position_indices
    )
    reward_curves.append(avg_reward)
    # Build curve for average X position
    avg_x = avg_position_stats[0]["avg"]
    avg_y = avg_position_stats[1]["avg"]
    position_curves.append({
        "x": steps,
        "y": avg_x,
        "label": "Avg X Pos"
    })
    position_curves.append({
        "x": steps,
        "y": avg_y,
        "label": "Avg Y Pos"
    })

# Group x/y values by label for plotting
plot_ready_data = {}
print("position_curves\n", position_curves)
for point in position_curves:
    label = point["label"]
    if label not in plot_ready_data:
        plot_ready_data[label] = {"x": [], "y": [], "label": label}
    plot_ready_data[label]["x"].append(point["x"])
    plot_ready_data[label]["y"].append(point["y"])

# Final plotting
print("plot_ready_data\n", plot_ready_data,'\n\n')
print("plot_ready_data.values\n", plot_ready_data.values(),'\n\n')
print("list(plot_ready_data.values)\n", list(plot_ready_data.values()),'\n\n')
curves_list = list(plot_ready_data.values())
print("curves_list\n", curves_list,'\n\n')
plot_position_curves_with_note(
    curves=curves_list,
    y_label="Avg X Position (per eval)",
    title="Avg X Position vs Training Progress",
    notes="Tracked average x-position over 5 episodes at each checkpoint"
)
'''plot_position_curves_with_note(
    curves=curves_list,
    y_label="Avg Reward (per eval)",
    title="Avg Reward vs Training Progress",
    notes="Tracked average reward over 5 episodes at each checkpoint"
)'''

```
